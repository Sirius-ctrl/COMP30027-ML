```python
from sklearn.metrics import confusion_matrix
def conf_matrix(predictor, dataset, l, detail=False):
    lb = ["Perth", "Melbourne", "Sydney", "Brisbane"]
    confu=confusion_matrix(l,
                 predictor.predict(dataset),
                 labels=lb)
    if detail:
        print(confu)
        
    for i in range(len(confu)):
        print(lb[i], confu[i][i]/np.sum(confu[i]), end=" ")
        worst = [v for v in confu[i]]
        worst[i] = 0
        worst_arg = np.argmax(worst) 
        print("worst =", lb[worst_arg], confu[i][worst_arg]/np.sum(confu[i]))
```

```python
from sklearn.metrics import confusion_matrix
def conf_matrix_one(predictions, l, detail=False):
    lb = ["Perth", "Melbourne", "Sydney", "Brisbane"]
    confu=confusion_matrix(l,
                 [idx_reverse[i] for i in predictions],
                 labels=lb)
    if detail:
        print(confu)
        
    for i in range(len(confu)):
        print(lb[i], confu[i][i]/np.sum(confu[i]))
```

```python
def word_to_index(oh):
    res = []
    for i in range(len(oh)):
        res += oh[i]*[i]
    return res
```

```python
def prepare_raw(raw_data):
    texts = []
    labels = []
    for i in range(len(raw_data)):
        labels.append(raw_data.iloc[i][1])
        texts.append(raw_data.iloc[i][2][1:-1])
    return texts, labels
```

```python
import re
import string

regex = re.compile('[%s]' % re.escape(string.punctuation))

def stringbreaker(text, num=True):
    text.translate(str.maketrans('', '', string.punctuation))
    sp = text.split(" ")
    http = []
    res = []
    for i in sp:
        if "http" in i:
            http.append(i)
            continue
            
        if(len(i)>0):
            if i[0] == "@" or i[0] == "#":
                temp = i
                temp_proc = re.findall('[A-Z][^A-Z]*', temp)
                single = ""
                
                for t in temp_proc:
                    if len(t) != 1:
                        if len(single) != 0:
                            res.append(single)
                            res.append(t)
                            single = ""
                        else:
                            res.append(t)
                    else:
                        single += t
                        
                if len(single) != 0:
                    res.append(single)
                    single = ""
                    
                if len(temp_proc) == 0:
                    res.append(temp)
            elif i[0] == '\\':
                #remove emoji
                pass
            else:
                res.append(i)
    # no enlongated
    #res = [re.sub(r'(.)\1{2,}', r'\1', remove_punc(i).lower()) for i in res if "|"+i.lower()+"|" not in swords]
    # has enlongated
    res = [remove_punc(i).lower() for i in res if "|"+i.lower()+"|" not in swords]
    
    if num:
        return remove_number(" ".join(res))
    #res += http
    return " ".join(res)

def remove_punc(s):
    return regex.sub("",s)

def remove_repet(a):
    i = len(a)-1
    while i > 1:
        if a[i] == a[i-1]:
            i-=1
        else:
            return a[:i+1]
    return a[:i+1]

def remove_number(s):
    return re.sub(r'[0-9]+', '', s)
```

```python
def nn_batch_generator(X_data, y_data, batch_size):
    samples_per_epoch = X_data.shape[0]
    number_of_batches = samples_per_epoch/batch_size
    counter=0
    index = np.arange(np.shape(y_data)[0])
    while 1:
        index_batch = index[batch_size*counter:batch_size*(counter+1)]
        X_batch = X_data[index_batch,:].todense()
        y_batch = y_data[index_batch]
        counter += 1
        yield np.array(X_batch),y_batch
        if (counter > number_of_batches):
            counter=0

model_raw.fit_generator(generator=nn_batch_generator(train_word_features, yr_train, 128),
                    nb_epoch=15, 
                    samples_per_epoch=train_word_features.shape[0],
                    verbose=1
                   )
```

```python
def plot_history(history):
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs = range(1, len(loss)+1)
    
    plt.plot(epochs, loss, "r", label="Training loss")
    plt.plot(epochs, val_loss, "b", label="Validation loss")
    plt.title("Training and validation loss")
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    
    plt.plot(epochs, acc, "r", label="Training acc")
    plt.plot(epochs, val_acc, "b", label="Validation acc")
    plt.title("Training and validation acc")
    plt.xlabel('Epochs')
    plt.ylabel('acc')
    plt.legend()
    plt.show()
    
```
